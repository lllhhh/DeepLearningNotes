# 1.快速绘图

## 1.1使用pyplot模块绘图

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 1000)
y = np.sin(x)
z = np.cos(x**2)

plt.figure(figsize=(8,4))

plt.plot(x, y, label="$ sin(x) $", color= 'red', linewidth=2)
plt.plot(x, z, 'b--', label='$ cos(x^2) $')
plt.xlabel("Time(s)")
plt.ylabel("Volt")
plt.title("example")
plt.ylim(-1.2, 1.2)
plt.legend()
plt.show()
```

```python
plt.figure(1)
plt.figure(2)
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

x = np.linspace(0, 3, 100)
for i in range(1,5):
    plt.figure(1)
    plt.plot(x, np.exp(i*x/3))
    plt.sca(ax1)
    plt.plot(x,np.sin(i*x))
    plt.sca(ax2)
    plt.plot(x, np.cos(i*x))
```
