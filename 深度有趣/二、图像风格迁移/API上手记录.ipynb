## keras.concatenate 矩阵连接

```python
import numpy as np
import keras.backend as K
import tensorflow as tf

a = K.variable(np.array([[1,2,3]]))
b = K.variable(np.array([[3,2,1]]))
c1 = K.concatenate([a, b], axis=0)
c2 = K.concatenate([a, b], axis=1)

init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    print(sess.run(c1))
    print(sess.run(c2))
```

# np.clip 截取函数
> 作用：将范围外的数强制转化为范围内的数
> def clip(a, a_min, a_max, out=None):
参数：
- a_min: 最小值
- a_max:最大值
- out: 把强制截取后的结果放到这个数组中，但是out中的数组必须和a形状一样

```python
a = np.arange(10)
a = np.clip(a=a, a_min=2, a_max=7, out=None)
a
```
