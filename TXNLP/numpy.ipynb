## 数组的转置和轴兑换

```python
arr = np.arange(15).reshape((3,5))
print(arr)
print(arr.transpose())
print(arr.T)
```

```python
arr = np.arange(24).reshape((2,3,4))
print(arr)
print(arr)
print(arr.transpose((1,2,0)))
```

### 值替换

```python
arr = np.random.randn(4,4)
print(arr)
r = np.where(arr>0 , 2, -2)
print(r)
arr_2 = np.where(arr>0, 2, arr)
print(arr_2)
```

### numpy数学运算
常用的sum mean std

```python
arr = np.random.randn(4,4)
print(arr)
print('mean：',arr.mean())
print('sum：',arr.sum())
print("std：",arr.std())
```

### 在一个轴上进行运算

```python
print(arr.mean(axis=1)) #计算轴上1的平均值
print(arr.sum(0)) #计算轴0上的和
```

### numpy文件操作
numpy可以读写磁盘上的文本数据或者是二进制数据 主要用用的函数就是np.save和np.load,
默认情况下保存的是原始的二进制格式保存在扩展名为.npy的文件中

```python
arr = np.arange(10)
np.save('any_array',arr)
np.load('any_array.npy')
```

```python
arr = np.arange(10)
np.savez('any_array_1',a=arr)
np.load('any_array_1.npz')['a']
```

```python
arr = np.arange(10)
np.savetxt('any_array.txt', arr, delimiter=',')
np.loadtxt('any_array.txt', delimiter=',')
```

### 随机漫步

```python
# 求什么时候第一次，距离初始点10步远
position = 0 #初始化位置
walk = [position]# 步值初始化
steps = 1000 #步数设置
for i in range(steps):
    step = 1 if np.random.randint(0,2) else -1
    position += step
    walk.append(position)

import matplotlib.pyplot as plt
plt.plot(walk)
print("abs：",np.abs(walk))
print('>10:',np.abs(walk)>10)
# 求什么时候第一次，距离初始点10步远
print((np.abs(walk)>10).argmax())
# 求最大值索引
print(np.abs(walk).argmax())
```
